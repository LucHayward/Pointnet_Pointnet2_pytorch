```python
def convert_to_df(data):
    """
    Converts data from a string of items on newlines to a 13 column df
    :param data: a string containing data
    :return: a Pandas DataFrame with 13 columns and the first 13 lines as headers
    """
    lines = data.strip().split('\n')
    headers = lines[:13]
    rows = [lines[i:i+13] for i in range(13, len(lines), 13)]
    df = pd.DataFrame(rows, columns=headers)
    return df
```
