```python
import pandas as pd


def convert_to_df_tabs(data):
    """
    Converts data from a tab-separated values (TSV) string to a Pandas DataFrame
    :param data: a TSV string with two rows of data
    :return: a Pandas DataFrame with the two rows of data as rows and the first row as column headers
    """
    lines = data.strip().split('\n')
    rows = [line.split('\t') for line in lines]
    df = pd.DataFrame(rows[1:], columns=rows[0])
    return df


def convert_to_df_csv(data):
    """
    Converts data from a comma-separated values (CSV) string to a Pandas DataFrame
    :param data: a CSV string with two rows of data
    :return: a Pandas DataFrame with the two rows of data as rows and the first row as column headers
    """
    lines = data.strip().split('\n')
    rows = [line.split(',') for line in lines]
    df = pd.DataFrame(rows[1:], columns=rows[0])
    return df
```

```python
import matplotlib.pyplot as plt

# Compute the average of column 'intensity_diff'
avg_intensity_diff = df['intensity_diff'].mean()
print(f"Average mIoU change by adding intensity {avg_intensity_diff * 100:.2f}%")

# Plot the DataFrame as a bar plot
ax = df.plot(x='log_dir', y='intensity_diff', kind='bar')

# Add an average line to the plot
ax.axhline(avg_intensity_diff, color='red')

# Show the plot
plt.show()
```

```python
import matplotlib.patches as mpatches
import matplotlib.lines as mlines

def fix_labels(label, index):
    """
    Fix labels by changing s3dis -> S3DIS and removing 'final/' and the name of the dataset
    :param label: the label to shorten
    :return: the shortened label
    """
    label = label.replace('s3dis', 'S3DIS').replace('final/', '').replace('_', ' ')

    if index == 0: label = label[13:]
    elif index ==1: label = label[7:]
    elif index == 2: label = label[10:]
    else: label = label[9:]

    if 'S3DIS' in label:
        label = label.replace('S3DIS', '')
        label = 'Pretrained ' + label

    return label

# Create a 2x2 grid of subplots with increased height and spacing between subplots
fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(20, 4))

# Loop through each subplot and plot the corresponding data
for i, ax in enumerate(axs.flatten()):
    start_idx = i * 8
    end_idx = start_idx + 8
    subset_df = df.iloc[start_idx:end_idx]
    subset_df.plot(x='log_dir', y='intensity_diff', kind='bar', ax=ax,
                   color=[x for x in ['tab:blue', 'tab:cyan'] for i in range(4)])
    # Compute the average of column 'intensity_diff'
    avg_intensity_diff = subset_df['intensity_diff'].mean()
    print(f"Average/Max/Min mIoU change by adding intensity {avg_intensity_diff * 100:.2f}/{subset_df['intensity_diff'].max()* 100:.2f}/{subset_df['intensity_diff'].min()* 100:.2f}%")

    ax.axhline(avg_intensity_diff, color='red')
    ax.set_title(f"{['Bagni di Nerone', 'Church', 'Montelupo', 'Monument'][i]}")
    # Rotate the x-axis labels by 45 degrees
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
    ax.legend().set_visible(False)
    if i == 0: ax.set_ylabel('mIoU difference')
    ax.set_xticklabels(subset_df['log_dir'].apply(lambda x: fix_labels(x, i)))
    ax.set_xlabel('')
    ax.set_ylim([-0.2, 0.2])
    ax.tick_params(axis='x', which='both', length=0)
    ax.set_xticks([])

pretrained_patch = mpatches.Patch(color='tab:blue', label='Pretrained')
random_patch = mpatches.Patch(color='tab:cyan', label='Random Init')
average_line = mlines.Line2D([], [], color='red', label='Average Difference')

handles = [pretrained_patch, random_patch, average_line]
fig.legend(handles=handles, bbox_to_anchor=(1.04, .9), loc='upper right')

# Adjust the spacing between subplots
# fig.subplots_adjust(hspace=1)
fig.subplots_adjust(wspace=0.3)

# Show the plot
plt.savefig('intensity_diff.pdf', bbox_inches='tight')
plt.show()
```
